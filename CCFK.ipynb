```python
data1= data.sample(frac = 0.1,random_state=1)

data1.shape
```

```python
data.shape
```

```python
Fraud = data1[data1['Class']==1]

Valid = data1[data1['Class']==0]

outlier_fraction = len(Fraud)/float(len(Valid))

print(outlier_fraction)

print("Fraud Cases : {}".format(len(Fraud)))

print("Valid Cases : {}".format(len(Valid)))
```

```python
## Correlation
import seaborn as sns

#get correlations of each features in dataset
corrmat = data1.corr()
top_corr_features = corrmat.index
plt.figure(figsize=(25,25))

#plot heat map
g=sns.heatmap(data[top_corr_features].corr(),annot=True,cmap="RdYlGn")
```

```python
columns = data1.columns.tolist()

columns = [c for c in columns if c not in ["Class"]]

target = "Class"

state = np.random.RandomState(42)
X = data1[columns]
Y = data1[target]
X_outliers = state.uniform(low=0, high=1, size=(X.shape[0], X.shape[1]))

print(X.shape)

print(Y.shape)
```

```python
#Prediction Model
#Isolation Forest, Local Outlier Factor used but not Support Vector Machine Algorithms
```

```python
##SVM 'random_state' issue (solution - https://scikit-learn.org/stable/modules/generated/sklearn.svm.OneClassSVM.html)

classifiers = {
    "Isolation Forest":IsolationForest(n_estimators=100, max_samples=len(X), contamination=outlier_fraction, random_state=state, verbose=0),
    "Local Outlier Factor":LocalOutlierFactor(n_neighbors=20, algorithm='auto', leaf_size=30, metric='minkowski', p=2, metric_params=None, contamination=outlier_fraction),
}
```

```python
#Should have been:- 
#"Support Vector Machine":OneClassSVM(kernel='rbf', degree=3, gamma=0.1,nu=0.05, max_iter=-1, random_state=state)
```

```python
type(classifiers)
```

```python
n_outliers = len(Fraud)
for i, (clf_name,clf) in enumerate(classifiers.items()):
    
    if clf_name == "Local Outlier Factor":
        y_pred = clf.fit_predict(X)
        scores_prediction = clf.negative_outlier_factor_
    else:    
        clf.fit(X)
        scores_prediction = clf.decision_function(X)
        y_pred = clf.predict(X)
    
    #0 for Normal and 1 for Fraud Transactions
    
    y_pred[y_pred == 1] = 0
    y_pred[y_pred == -1] = 1
    n_errors = (y_pred != Y).sum()
    
    # Run Classification Metrics
    print("{}: {}".format(clf_name,n_errors))
    print("Accuracy Score :")
    print(accuracy_score(Y,y_pred))
    print("Classification Report :")
    print(classification_report(Y,y_pred))
```

```python
#### Results :

- Isolation Forest detected 73 errors and Local Outlier Factor detected 97 errors.
- Accuracy of Isolation Forest is 99.74% whereas Local Outlier Factor is 99.65%.
- Error Precision & Recall of the Isolation Forest is 27% and Local Outlier Factor is just 2%.
- Overall, Isolation Forest Method performed better as compared to Local Outlier Factor in determining the fraud cases by 25%.
- The accuracy of the fraud detection can be improved via Deep Learning Algorithums or by using bigger sample size but both of them has a computational cost.
- Complex Anomaly Detection Models like Bayesian Network and Statistics can also be used to have a higher accuracy in determining fraudulent cases.
```
